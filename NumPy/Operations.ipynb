```python
import numpy as np
```

```python
# defining 2 nd arrays
arr1 = np.array([1,2,3,4,5,6])
arr2 = np.array([4,5,6,7,8,9])
```

```python
# element wise operations
print(arr1-arr2)
print(arr1*arr2)
print(arr1/arr2)
print(arr1+arr2)
```

```python
# we can perform operations on single array also 
print(arr1*2)
print(arr1+1)
```

```python
# checks condition for every element 
arr1>3
```

```python
# we can also do the dot product of 2 array
arr3=np.arange(0,10).reshape(5,2)
arr4=np.arange(10,20).reshape(2,5)
arr3.dot(arr4)
```

```python
# to find maximum or minimum element
print(arr4.max())
print(arr4.min())
```

```python
# to print collumn and row, max and min elements
arr4.max(axis = 0)#collumn max
arr4.max(axis = 1)#row max
```

```python
# to find sum of all elements
arr1.sum()
```

```python
#to find sum of elements in all collumn or row
print(arr4.sum(axis=0))#collumn
print(arr4.sum(axis=1))#row
```

```python
# to find mean of a array
arr4.mean()
```

```python
# to calculate standard deviation
arr4.std()
```
