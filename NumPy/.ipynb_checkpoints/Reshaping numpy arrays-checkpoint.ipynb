```python
import numpy as np
```

```python
arr1 = np.arange(10).reshape(2,5)
arr1
```

```python
# ravel funtion can make any array into 1d
arr1.ravel()
```

```python
# reshape we already know
```

```python
# we can find transpose of array
arr1.transpose()
```

```python
# we can combine two arrays of same shape 
arr2 = np.arange(10,20).reshape(2,5)
print(np.hstack((arr1,arr2)))
print(np.vstack((arr1,arr2)))
```

```python
# we can split also 
print(np.vsplit(arr1,2))
print(np.hsplit(arr1,5))
```
