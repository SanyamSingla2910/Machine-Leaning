```python
#importing numpy 
import numpy as np
```

```python
# creating a 3d array
arr1 = np.array([[[3,4],[1,2]],[[5,6],[7,8]]])
# tells us the edges of the dimension of the array
arr1.shape
```

```python
# getting dimension of that arary
arr1.ndim
```

```python
#size gives the number of elements of that array
print(arr1.size)
#itemsize tells the memory occupied by single item of the array  
print(arr1.itemsize)
```

```python
#dtype tells the datatype of elements inside it 
arr1.dtype
```

```python
#to change datatype to float
arr1 = arr1.astype(float)
```

```python
# checking data type
arr1.dtype
```
