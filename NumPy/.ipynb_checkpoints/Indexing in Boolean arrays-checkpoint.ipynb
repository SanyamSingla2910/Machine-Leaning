```python
import numpy as np
```

```python
# we used randint to create a 
arr1 = np.random.randint(1,100,(6,5))
arr1
```

```python
# this will create a boolean array
(arr1>50)
```

```python
# to get the elements that are greater than 50 we use
arr1[arr1>50]
```

```python
# we can do multiple thing using this 
print(arr1[arr1%10==0])
print(arr1[(arr1>=25) & (arr1%2==0)])
```

```python
# we can use it to change values of that numbers also 
arr1[(arr1>=25) & (arr1%2==0)]=0
arr1
```
