```python
import numpy as np
```

```python
arr1 = np.random.randint(0,50,(3,3))
arr2 = np.random.randint(0,50,(1,3))
print(arr1)
print(arr2)
```

```python
# we can see here that we can apply arithmetic operations on diferent shape of arrays also 
print(arr2/arr1)
print(arr2*arr1)
print(arr2+arr1)
print(arr2-arr1)
```

```python
# now we will check different combinations that we can do in a brodcasting
arr1 = np.arange(1,2).reshape(1,1)
arr2 = np.arange(100,200).reshape(5,20)
# if the array has 1 by 1 dimension then it can be brodcasted with any array
arr1*arr2
```

```python
arr1 = np.arange(0,5).reshape(5,1)
arr2 = np.arange(5,10).reshape(1,5)
arr1*arr2
```

we just have to visualize that the arrays can be brodcasted or not just simple as that
