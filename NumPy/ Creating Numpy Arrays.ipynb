```python
#creating numpy/N-D arrays
import numpy as np
```

```python
arr1 = np.array([1,2,3,4,5,6])
arr1
```

```python
 # to check the type of our arr1
type(arr1)
```

```python
# printing a 2d array
arr2 = np.array([[1,2,3],[4,5,6]])
arr2
```

```python
# creating array with all elements as zero 
arr3 = np.zeros((2,3))
arr3
```

```python
# creating array with all elements as one
arr4 = np.ones((2,3))
arr4
```

```python
# creating identity martix 
arr5 = np.identity(5)
arr5
```

```python
#creating array from 0 to a number
arr6 = np.arange(15)
print(arr6)
#creating from a differnt stating number 
arr7 = np.arange(5,11)
print(arr7)
# we can also tell how much gaps should be between them
arr8 = np.arange(10,21,2)
print(arr8)
```

```python
#creating array with equidistant points in them
arr9 = np.linspace(0,20,9)
arr9
```

```python
# copying an array
arr10 = np.copy(arr9)
arr10
```
