```python
import numpy as np
```

```python
# using random function
arr1 = np.random.random()
arr1
```

```python
# this gives the same random number again and again
np.random.seed(1)
arr1 = np.random.random()
arr1
```

```python
# to create random float numbers within a range
arr1 = np.random.uniform(1,11,10).reshape(5,2)  #we can also reshape array
arr1
```

```python
# to generate integer values we can use this
arr1 = np.random.randint(0,100,30).reshape(6,5)
arr1
```

```python
#  to get index of max or min values 
print(np.argmax(arr1))
print(np.argmin(arr1))
```

```python
#  we can check also if that is correct by indexing 
arr1 = np.ravel(arr1)
arr1[np.argmax(arr1)]
```

```python
# we can chenge values as we need also 
arr1[arr1%2==0]= 3
arr1
```

```python
# if we want to change elemewnts as we need but dont want to change original array
arr1 = np.random.randint(0,100,30).reshape(6,5)
print(np.where(arr1%2==0,1,-1))
print(arr1)
```

```python
# we can sort the array elements also 
arr1 = np.ravel(arr1)
arr1 = np.sort(arr1)
```

```python
# we can use the percentile funtion also to get the value as we need
np.percentile(arr1,29.45)
```
