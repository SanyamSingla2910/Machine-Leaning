```python
# pandas is the most useful library that is used to work with data
import pandas as pd
```

```python
# we can import data from our system locally
df = pd.read_csv('train_and_test2.csv')
df.head(5)
```

```python
# sep parameter is used so we can seperate values with any other thing than tsv
# here for example i have a tsv dataset 
pd.read_csv('TSV.tsv',sep ='\t')
# i have to set sep parameter as \t so that it can read tsv files 
```

```python
# if i don't have any name of our collumns we can pass a list of names so that we can easily setup names for our collumns
pd.read_csv('TSV.tsv',sep='\t',names = ['ID','Name','Age','Subject','Numbers']).iloc[1:,:]
# as you can see i removed the original collumn and set it by myself 
```

```python
# we can set index collumn to any other collumm 
pd.read_csv('TSV.tsv',sep = '\t',index_col='StudentID')
```

```python
# if i dont want first 5 rows i  can use the header parameter 
pd.read_csv('train_and_test2.csv',header=6)
```

```python
# i want only hours studied and marks so i can use usecols to select cols
pd.read_csv('Students.csv',usecols = ['StudentID','Hours_Studied','Marks'],index_col='StudentID')
```

```python
# if we want to skip some rows or want some particular rows we can use this function
pd.read_csv('Students.csv',skiprows=[0,1,2,4,5,6,7,8,9,10])
```

```python
# if we want less data we can use nrows so that we select only the rows we want 
pd.read_csv('Students.csv',nrows = 10)
```

```python
# to skip the bad lines
pd.read_csv('Students.csv',on_bad_lines='skip')
```

```python
# we can use dtype function to change the datatype of the features
pd.read_csv('Students.csv',dtype={'Marks':float,'Attendance':float})
```

```python
# we can use parse_dates to set a collumn as dates collumn
pd.read_csv('C:/Skills/Machine Learning/Pandas/IPL MATCHES.csv',parse_dates=['date'])
```

```python
# we can use converters which will pass a function and the collumn that we want to apply the function on
def debar(att):
    att = int(att) 
    if att>=75:
        return att
    else:
        return "Debared"
pd.read_csv('Students.csv',converters={'Attendance':debar})
```

```python
# example we dont need the data of 75 so what we can do is 
pd.read_csv('Students.csv',na_values=[75] )
```

```python
# loading huge dataset in chunks
pd.read_csv('train_and_test2.csv')#there are 1308 rows 
dfs = pd.read_csv('train_and_test2.csv',chunksize=300)
for meow in dfs:
    print(meow)
# we can use for loops to acess the data divided in chunks 
```
