```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36"}
# this header is used so that the website thinks that we used browser to hit the url instead of the 
data = requests.get('https://en.wikipedia.org/wiki/List_of_highest-grossing_Indian_films',headers=h).text
```

```python
# now you can see that we fethched the url using requests but making the website think we used a browser 
# we will use beautifulSoup to pass our webpage 
soup = BeautifulSoup(data)
```

```python
print(BeautifulSoup.prettify(soup))
```

```python
# here i am finding all the data of hightest grossing top 50 movies 
item =  soup.find_all('td')  #we have to put html tag to scrape the data in that tag 
for i in item:
    print(i.text.strip())
```

```python
# like i need data for a particular table 
item = soup.find('table',class_ = 'wikitable')
for i in item:
    print(i.text.strip())
```

```python
# now for example we need get the rows of the table we extracted 
movies = []
row = item.find_all('tr')
for i in row:
    cols = (i.find_all('td'))
    for data in cols:
        movies.append(data.text.strip())
```

```python
# we created a list of all the data 
movies
```

```python
new_movies = []
for i in range(0,len(movies),5):
    new_movies.append(movies[i:i+5])
new_movies
```

```python
# we will fetch a data and convert it into a csv file which will make it more accessible for our pandas library 
# firstly we will fetch the data from the URL and convert it to text
data = requests.get('https://realpython.github.io/fake-jobs/').text
```

```python
type(data)
```

```python
# loading the html in beautiful soup 
html = BeautifulSoup(data)
```

```python
# to get every card details
card = html.find_all('div',class_ = 'card-content')
title = []
# to get the title 
for i in card:
    titles = (i.find_all('h2',class_ = 'title'))
    for j in titles:
        title.append(j.text.strip())
len(title)
```

```python
# to get companies 
companies = []
for i in card:
    titles = (i.find_all('h3',class_ = 'subtitle'))
    for j in titles:
        companies.append(j.text.strip())
len(companies)
```

```python
location = []

for i in card:
    titles = (i.find_all('p',class_ = 'location'))
    for j in titles:
        location.append(j.text.strip())
len(location)
```

```python
# now we have the location title of job and company now we can put them in a dataframe and export it to csv file 
df = pd.DataFrame({'Job_title':title,'Company_Name':companies,'Location':location})
```

```python
df
```

```python
# now we can create its csv file 
df.to_csv('Job Data.csv')
```
