```python
# i got info about him 
# now i want to fetch titles from the api
titles = requests.get('https://api.watchmode.com/v1/list-titles/?apiKey=Gv6A5xOxZcjI21EhKaMBGTTrMK20IyhevxHASHlS')
pd.DataFrame(titles.json()['titles'])
```

```python
df = pd.DataFrame()
```

```python
# to get data of all pages 
for i in range(1,50):
    titles = requests.get('https://api.watchmode.com/v1/list-titles/?apiKey=Gv6A5xOxZcjI21EhKaMBGTTrMK20IyhevxHASHlS&source_ids={}'.format(i))
    temp = pd.DataFrame(titles.json()['titles'])
    df = pd.concat([df,temp],ignore_index=True)
```

```python
#i used this to fetch data from multiple pages which will give me all data in every page which will give me a list of movies
df = df[['title','year']]
df.head()
```
