```python
# we can see there are no missing values 
nd = df.values
nd[0,3]=np.nan
```

```python
# used to convert nd array into dataframe
df = pd.DataFrame(nd)
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.rename(columns={0:'ID',1:'name',2:'age',3:'math',4:'science',5:'english',6:'attendance'})
df
```

```python
# filtering 
# student with only maths marks >80
df = df[df['math']>80]
df
```

```python
# df where attendance is 90+
df = df[df['attendance']>90]
df
```

```python
# student aged 19
df= df [df['age']==19]
df
```

```python
# sorting 
# by maths marks 
df.sort_values('math',inplace=True)
```

```python
# sort by attendance 
df.sort_values('attendance',ascending=False,inplace=True)
```

```python
df
```

```python
df['Total'] = df['math']+df['science']+df['english']
```

```python
df.drop('grade',axis=1,inplace=True)
```

```python
df
```

```python
df['percentage'] = (df['Total']/300)*100
```
