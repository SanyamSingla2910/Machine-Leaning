```python
import numpy as np
import pandas as pd
```

```python
match = pd.read_csv('IPL MATCHES.csv')
df = pd.read_csv('deliveries.csv')
```

```python
#merge
#we need to find list of orange cap holders = (most runs in a particular series)
new = df.merge(match,left_on = 'match_id',right_on = 'id')
```

```python
data= new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending = False).reset_index().drop_duplicates('season')
data.sort_values('season')
```
