```python
import numpy as mp
import pandas as pd
```

```python
df = pd.read_csv('deliveries.csv')
df.head(2)
```

```python
# Find the most destructive death over batsman in ipl history 
# he must've played 200 balls
```

```python
# no. of balls played by every player 
death_over = df[(df['over']>=16)]
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
x = all_batsman>200
batsman_list = all_batsman[x].index.tolist()
```

```python
mask = df['batsman'].isin(batsman_list)
final = df[mask]
```

```python
# now we will calculate runs and balls of this given dataframe
balls = final.groupby('batsman')['batsman_runs'].count()
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
sr = ((runs/balls)*100)
```

```python
sr.sort_values(ascending = False)
```
