```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('IPL MATCHES.csv')
```

```python
# drop duplicates is used to remove all the duplicates of the collumn we specify from the data
df.drop_duplicates(['city','season'])
# this will give us the matches that are played in different cities
```

```python
# we can use this to calculate winner of every ipl season
winners=df.sort_values('date',ascending = False).drop_duplicates('season')
winners[['season','winner']]
```
