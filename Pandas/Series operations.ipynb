```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('IPL MATCHES.csv')
```

```python
df.head()
```

```python
# if we perform a operation on series we get a series
type(df['winner'].value_counts())
```

```python
# we can also get the index or values by using some attributes
print(df['winner'].value_counts().index)
print(df['winner'].value_counts().values)
```

```python
# we can apply arithmetic operations on series with same indexes
#example we need to get number of matches played by every team 
def match_counter(team):
    matches_played = df['team1'].value_counts()+df['team2'].value_counts()
    print(matches_played[team])
match_counter('Mumbai Indians')
```
