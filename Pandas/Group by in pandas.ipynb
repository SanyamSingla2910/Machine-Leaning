```python
# now we will perfom questions on this dataset
# importing new data
data = pd.read_csv('deliveries.csv')
data.head()
```

```python
# get top 5 players who have most runs
batsman = data.groupby('batsman')
batsman['batsman_runs'].sum().sort_values(ascending=False).head(5)
```

```python
# to find batsman who scored most 4s
data_of_4s = data[data['batsman_runs']==4]
fours = data_of_4s.groupby('batsman')
```

```python
fours['batsman_runs'].count().sort_values(ascending=False)
```

```python
# to get most runs scored by vkohli in front of which team
batsman_runs = data[['batting_team','bowling_team','batsman','batsman_runs']]
batsman_runs
```

```python
VKohli_runs = batsman_runs[batsman_runs['batsman']=='V Kohli']
VKohli_runs
```

```python
Opp_team =VKohli_runs.groupby('bowling_team')
```

```python
Opp_team['batsman_runs'].sum().sort_values(ascending = False)
```

```python
# creating a funtion when we give player name to it then the function will return the team he scored max runs in oppose
def Max_runs_by_player_with_which_team(player):
    player_data = data[data['batsman']==player]
    return (player_data.groupby('bowling_team')['batsman_runs'].sum().sort_values().tail(1))
Max_runs_by_player_with_which_team("V Kohli")
```
