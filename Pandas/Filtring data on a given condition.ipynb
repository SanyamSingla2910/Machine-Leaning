```python
df = pd.read_csv('IPL MATCHES.csv')
```

```python
# this is also known as boolean series
mask = df['winner']=='Royal Challengers Bangalore'
mask
```

```python
# we can also get our dataframe or series on a complex condition
# example we only want data that where winner is rcb and match was player after 2010
df[(df['winner']=='Royal Challengers Bangalore') & (df['season']>2010)]
```

```python
# we can also create a function of them 
def winner_list_after_season(team,year):
    mask1 = df['winner']==team
    mask2 = df['season']>year
    return df[mask1 & mask2]

s = winner_list_after_season('Royal Challengers Bangalore',2010)
type(s)
```
