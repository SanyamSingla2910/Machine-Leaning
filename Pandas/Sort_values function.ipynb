```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("IPL MATCHES.csv")
```

```python
df.head()
```

```python
# we can sort values inside a dataframe or series using sort_values() function
(df['team1'].value_counts() + df['team2'].value_counts()).sort_values(ascending= False)
# we can also do ascending=False to make it in descending order
```

```python
# we can also use them on dataframes
df.sort_values('city',ascending = False ,inplace= True) #inplace is used to make changes in the actual data
```
