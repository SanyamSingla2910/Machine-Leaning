```python
# matplot lib is used to create visualizations which will help us understand our data more better
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Skills\Machine Learning\ML main\Understanding Data\Titanic-Dataset.csv')
df.head()
```

# Pie chart

```python
# this code is to get pie chart of something and we can show its percentage also 
Dead_Person = df['Survived'].value_counts()
pie_chart = plt.pie(Dead_Person,labels=['Not Survived','Survived'],autopct='%1.2f%%')
plt.show()
```

# Scatter Plot

```python
# to do scatter plot we need data and check relation between 2 data
plt.scatter(df['Age'],df['Fare'],color='blue')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title("Age vs Fare")
plt.show()
```

# Line Graph

```python
a=[100,200,300,400,500]
c=[2,44,65,8,7]
b=[1,2,23,4,34]
plt.plot(a,b,color = 'red',marker = '+',label='Red line')
plt.plot(a,c,color = 'yellow',marker = "d",label='Yellow Line')
plt.legend()
plt.show()
```

# Bar Graph 

```python
Year=[2010,2011,2012,2013]
Expense=[100,200,300,400]
plt.bar(Year,Expense)
plt.show()
```

# Histogram

```python
plt.hist(df['Age'])
plt.show()
```

# Box Plot

```python
plt.boxplot(df['Fare'])
plt.show()
```
